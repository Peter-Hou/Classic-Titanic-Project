At this step, my goal is to observe any columns with null values and identify the correct data types.

By observation, columns with null values are
- Age
- Cabin
- Embarked

The columns may have incorrectly identified data types are (note the following is based on the column definition given out in the competition setting)
- Categorical
    - Survived (Ordinal)
        - 1 = Yes
        - 0 = No
    - Pclass (Ordinal)
        - 1 = 1st
        - 2 = 2nd
        - 3 = 3rd
    - Sex (Nominal)
        - female
        - male
    - Embarked (Nominal)
        - C = Cherbourg
        - Q = Queenstown
        - S = Southampton
    - Name (Nomial)
- Numerical
    - PassengerId (Discrete)
    - Age (Continuous)
    - SibSp (Discrete)
    - Parch (Discrete)
    - Fare (Continuous)
- Others
    - Cabin (Alphanumeric or Categorical)
    - Ticket (Numeric or Alphanumeric)

Note that there are only 204 non-null values for Cabin and 147 of them are different number. Therefore, it is likely that one Cabin is used for one person with a few exceptions. I would treat it as categorical variable, but I do not think it is going to provide much useful correlation with 'Survived" hence will be dropped shortly.

### Visualizing the Correlation between categorical variable and "Survived"

#### for "Pclass"

```python
train[['Pclass', 'Survived']].groupby(['Pclass'], as_index = False).mean().sort_values(by = 'Survived', ascending=False)
```

The 1st class passenger has the highest rate of survival, followed by 2nd class passenger and the 3rd class passengers

#### for Sex

Female are most likely to survive.

#### Embarked

```python
train[['Embarked', 'Survived']].groupby('Embarked', as_index = False).mean().sort_values(by = 'Survived', ascending = False)
```

People embarked at port C are mostly likely to survive with people board at port Q and port S having about the same rate of survival.

Notably, there might not be many people board port C and Q in the dataset.

### Visualization Correlation between Numeric Variable and 'Survived'

```python
def hist_plot_numeric_corr(numeric_var):
    plt.figure(figsize = (6,4))
    sns.histplot(data = train, x = numeric_var, hue = 'Survived',  
                 palette = sns.color_palette(['green', 'orange']), multiple = 'stack').set_title(f"{numeric_var} vs Survived")
    plt.show()                                           
    
    
def bar_plot_numeric_corr(numeric_var):
    plt.figure(figsize = (6, 4))
    sns.barplot(data = train, x = numeric_var, y = 'Survived').set_title(f"{numeric_var} vs Survived")
    plt.show()
```

#### Age

```python
sns.set_style('dark')
hist_plot_numeric_corr('Age')
```

- Infants of age(0 - 5), Children of age(10-15), the elderly (> 75), and other passengers of age (15-40) are most likely to survive

#### SibSp

```python
bar_plot_numeric_corr('SibSp')
```

#### Parch

```python
bar_plot_numeric_corr('Parch')
```

### Using Correlation Plot for Numerical Variable

```python
sns.set(rc = {'figure.figsize': (10, 6)})
sns.heatmap(train[['Survived', 'Age', "SibSp", 'Parch', 'Fare']].corr(), annot = True, fmt = '.2g')
```

Key Results:
- Fare (0.26) exhibited a much higher correlation with Survived than the other features
- Age negatively correlated with SibSp(-0.31) and Parch(-0.19)
    - meaning that the lower the age, the more siblings and parents are on board at the same time.
- SibSp and Parch are positively correlated with a highest correlation coefficient in the above plot at 0.41.

Insights:
- SibSp and Parch could be combined together as one feature standing for the number of relatives on board.

### Interaction between categorical and numerical variables

```python
def plot_interaction_betwen_categorical_numerical(category, numeric):
    grid = sns.FacetGrid(train, col = 'Survived', row = category,
    size = 2.2, aspect = 1.6)
    grid.map(plt.hist, numeric, alpha = 0.5, bins = 20)
    grid.add_legend
    
categories = ['Pclass', 'Sex', 'Embarked'] * 4
numeric =  ['Age'] * 3 + ['SibSp'] * 3 + ['Parch'] * 3 + ['Fare'] * 3

for category, numeric in zip(categories, numeric):
    plot_interaction_betwen_categorical_numerical(category, numeric)
```

```python

train.Embarked.fillna(train.Embarked.dropna().max(), inplace = True)

for dataset in combine:
    dataset['Embarked'] = dataset['Embarked'].dropna().map({'S':0, 'C':1, 'Q':2}).astype(int)
```
